# **What is Exception Handling?**

**Sometimes your code has a mistake while it's running, like dividing by zero or using the wrong input. This mistake is called an exception.**

If you don't handle it, your program will crash.
But with exception handling, you can catch the error and show a friendly message instead.

image.png

# **Key Keywords:**

**try**: Block of code to test for errors.

**except**: Block to handle the error.

**finally**: (Optional) Always runs, even if there's an error.

# **Basic Syntax:**

```python
try:
    # Code that may raise an exception
    x=10
    y=0
    z=x/y
    print(z)

except ZeroDivisionError:
    # Code to handle the exception
    print("You can't divide by zero!")
finally:
    # Optional: Code that always runs
    print("This will always execute.")
```

# **Sample Code--TypeError:**

```python
# TypeError
try:
    result = "2" + 2
except TypeError:
    print("Wrong type used! You can't add a string and an int.")
```

## **Why we use multiple except blocks?**

**If You may want to handle different types of exceptions in different ways, all coming from the same try block.**

#**Syntax Example:**

```python
try:
    # Some risky code
    value = int(input("Enter a number: "))
    result = 10 / value
    print("Result:", result)
except ValueError:
    print("You entered a non-number! Please enter a valid number.")
except ZeroDivisionError:
    print("You entered zero! Cannot divide by zero.")
except Exception as e:
    print("Some other error occurred:", e)
```

# **Can we use multiple try blocks with one except block?**

**No, Python does not allow multiple try blocks to share a single except block directly.**

Each try must be followed by its own except block (or blocks).

# **What is raise?**

**The raise keyword in Python is used to manually trigger an error.we use it when you want to throw an exception on purpose in your code.**

You can raise built-in errors (like ValueError, KeyError) or create your own custom errors.

# **Example Code for raise:**

```python
# Example 1: Raising a built-in error (ValueError)
def check_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative!")  # Raise an error if age is negative
    else:
        print(f"Your age is {age}")

# Test the function
try:
    check_age(-5)
except ValueError as e:
    print(f"Caught an error: {e}")
```

# **What is the role of else in exception handling?**

The else block is executed if no exceptions are raised in the try block.

It is useful for code that should run only when no exceptions occur.

# **Syntax of try-except-else:**

```python
try:
    # Code that may raise an exception
    value = 10 / 2
except ZeroDivisionError:
    # Code that runs if an exception is raised
    print("You can't divide by zero!")
else:
    # Code that runs if no exception is raised
    print("The operation was successful!")
```
