# **What is Clustering?**

Clustering is an unsupervised learning technique used to group similar data points together based on their features. The goal is to identify patterns or structures in the data without predefined labels.

# **What is K-Means Clustering?**

K-Means Clustering is a machine learning algorithm used to group similar data points into clusters. It is an unsupervised learning technique, meaning it does not require labeled data.

**Example: Clustering Employees by Salary**

Suppose we have a list of employee salaries, and we want to group them into Low Salary, Medium Salary, and High Salary categories. K-Means will automatically divide them into k clusters based on patterns in the data.

# **What is DBSCAN Clustering?**

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a clustering algorithm that groups data points based on density rather than distance from a centroid (like K-Means).

Unlike K-Means, DBSCAN does not require you to specify the number of clusters (k) beforehand. It automatically finds clusters based on how close points are to each other.

# **1️⃣ What is ε (Epsilon) = 100?**

Epsilon (ε) is the maximum distance allowed for two customers to be considered neighbors.

If the total amount spent difference or visit frequency difference between two customers is ≤ 100, they are considered similar (neighbors).

**Example**:

If Customer A spends 500 and Customer B spends 520, the difference is 20, which is within ε = 100, so they are neighbors.

But if Customer C spends 1600, the difference from 500 is 1100, which is > 100, so they are NOT neighbors.

# **2️⃣ What is MinPts = 3?**

MinPts (Minimum Points) is the minimum number of similar customers (neighbors) needed to form a cluster.

If a customer has at least 2 other customers (total 3 including itself) within ε, it becomes a core point, forming a cluster.

If a customer has fewer than MinPts (less than 3 neighbors), it may become a border point or noise.

# **📌 Conclusion for DBSCAN Clustering of Customer Shopping Patterns**

Based on DBSCAN clustering, customers have been grouped into three segments based on their spending behavior and shopping frequency:

✅**Cluster 1 → Low Spenders (Occasional Buyers):**

These customers make infrequent purchases and spend small amounts.
Businesses can target them with discount offers to encourage more frequent shopping.

✅ **Cluster 2 → Regular Shoppers (Moderate Spend):**

These are loyal customers who shop frequently with a moderate budget.
Businesses should focus on loyalty programs, membership discounts, or personalized recommendations to retain them.

✅ **Cluster 3 → Luxury Shoppers (High-Value, Less Frequent Buyers):**

These customers make expensive but less frequent purchases.
Businesses can engage them with exclusive offers, premium product launches, and personalized services to increase spending.
