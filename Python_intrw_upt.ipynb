# **write a program for Binary Search?**

````python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
print(binary_search([1, 2, 3, 4, 5], 3))  # 2
````

# **Find the Maximum Subarray Sum (Kadane’s Algorithm)**

````python
def max_subarray_bruteforce(arr):
    max_sum = float('-inf')  # Smallest possible value

    for i in range(len(arr)):
        current_sum = 0
        for j in range(i, len(arr)):
            current_sum += arr[j]
            max_sum = max(max_sum, current_sum)

    return max_sum

print(max_subarray_bruteforce([-2, 1, -3, 4, -1, 2, 1]))  # Output: 6
````

# **write a program for Longest Palindromic Substring?**

````python
def is_palindrome(s):
    return s == s[::-1]  # Check if string is the same when reversed

def longest_palindrome_bruteforce(s):
    n = len(s)
    longest = ""

    for i in range(n):  # Start of the substring
        for j in range(i, n):  # End of the substring
            substring = s[i:j+1]  # Extract substring
            if is_palindrome(substring) and len(substring) > len(longest):
                longest = substring  # Update longest palindrome

    return longest

print(longest_palindrome_bruteforce("babad"))  # Output: "bab" or "aba"
````

# **write a program to Move Zeroes?**

````python
def move_zeroes(nums):
    non_zero_pos = 0
    for i in range(len(nums)):
        if nums[i] != 0:
            nums[non_zero_pos], nums[i] = nums[i], nums[non_zero_pos]
            non_zero_pos += 1
    return nums
print(move_zeroes([0, 1, 0, 3, 12]))  # [1, 3, 12, 0, 0]
````

# **Reverse an array in-place.**

````python
def reverse_array(arr):
    l= 0
    r=len(arr) - 1
    while l < r:
        arr[l], arr[r] = arr[r], arr[l]
        l += 1;
        r -= 1
    return arr
print(reverse_array([1, 2, 3, 4]))  # [4, 3, 2, 1]
````

**Example 9: Find the Missing Number in an Array**

````python
def find_missing_number(arr, n):
    # Step 1: Calculate the expected sum of numbers from 1 to n
    expected_sum = 0
    for i in range(1, n + 1):
        expected_sum += i

    # Step 2: Calculate the actual sum of array elements
    actual_sum = 0
    for num in arr:
        actual_sum += num

    # Step 3: Find the missing number
    missing_number = expected_sum - actual_sum
    return missing_number

# Example Usage
arr = [1, 2, 4, 5, 6]  # Missing number is 3
n = 6  # The range is from 1 to 6
print("Missing Number:", find_missing_number(arr, n))
````

# **write a Python program to count valid pairs in an array where the sum of the pair exists in the array**

Screenshot 2025-03-04 121812.png

````python
def count_valid_pairs(nums):
    # Initialize pair count
    pair_count = 0

    # Iterate through all unique pairs (i, j) where i < j
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            # Calculate the sum of the current pair
            pair_sum = nums[i] + nums[j]

            # Check if the sum exists in the list
            if pair_sum in nums:
                pair_count += 1

    return pair_count

# Example usage
nums = [2, 4, 3, 1, 10, 7, 6]
print("Total valid pairs:", count_valid_pairs(nums))  # Output: 7
````

# **Given a list of coins with different denominations, find the coin that contributes the highest total value (coin value × its count). If multiple coins have the same highest value, return the smallest coin among them.**

Screenshot 2025-03-04 123201.png

````python
from collections import Counter

def max_amount_coin(coin_list):
    # Count the frequency of each coin
    coin_count = Counter(coin_list)

    # Track the coin with the maximum amount
    max_amount = 0
    best_coin = 0

    # Iterate through each coin and calculate the total value
    for coin, count in coin_count.items():
        total_value = coin * count
        # Update if this coin gives a greater amount
        if total_value > max_amount:
            max_amount = total_value
            best_coin = coin

    return best_coin

# Example usage
coins = [1, 1, 1, 1, 1, 1, 5, 2, 2, 10, 1, 1, 1, 2, 1, 1, 1, 3, 4, 3]
print(max_amount_coin(coins))  # Output: 1
````

# **Given g grounds where each can host 2 teams, and t total teams, determine how many teams cannot play on the first day. If all teams can be accommodated, return 0, otherwise return the extra teams left out.**

Screenshot 2025-03-04 124143.png

````python
def solution(g, t):
    # Maximum number of teams that can play on the first day
    max_teams_playing = g * 2

    # If there are fewer teams than the capacity, no team is left out
    if t <= max_teams_playing:
        return 0

    # Otherwise, calculate how many teams won't play
    return t - max_teams_playing

# Example cases
print(solution(3, 10))  # Output: 4 (6 teams play, 4 are left out)
print(solution(5, 8))   # Output: 0 (All teams can play)
print(solution(2, 5))   # Output: 1 (4 teams play, 1 is left out)
print(solution(0, 6))   # Output: 6 (No grounds, all teams are left out)
print(solution(10, 10)) # Output: 0 (All teams can play)
````

# **write a program to print first non-repeated character in a given string?**

````python
def first_non_repeating_char(s):
    # Create a dictionary to store character counts
    char_count = {}

    # Count the frequency of each character
    for char in s:
        char_count[char] = char_count.get(char, 0) + 1

    # Find the first character with a count of 1
    for char in s:
        if char_count[char] == 1:
            return char

    return None  # Return None if no unique character is found

# Example usage
string = "abcdebaceg"
result = first_non_repeating_char(string)
print("First non-repeating character:", result)
````

# **write a program to generate a random password?**

````python
import random
import string

def generate_password(length):
    characters = string.ascii_letters + string.digits + string.punctuation  # Collection of all possible characters

    password_list = []  # Initialize an empty list to store selected characters
    for _ in range(length):
        random_char = random.choice(characters)  # Select a random character
        password_list.append(random_char)  # Add it to the list

    password = ''.join(password_list)  # Convert the list into a string
    return password

# Example usage
length = int(input("Enter password length: "))
password = generate_password(length)
print("Generated Password:", password)
````

# **what is mean by OOPS?**

**OOPS in Python stands for Object-Oriented Programming System. It is a programming style that organizes code using objects and classes to make it easier to manage and reuse.**

✅**Key Concepts of OOPS:**

**Class:** A blueprint/template for creating objects.

📌 Example: Think of a Car Blueprint – it defines how all cars should be made.

**Object:** An instance of a class (real-world entity).

📌 Example: A Toyota Car made from the car blueprint is an object.

**Attributes:** Variables inside a class (describe object properties).

📌 Example: Color, speed of a car.

**Methods:** Functions inside a class (define object behavior).

📌 Example: Drive, brake, honk.

#**Class**
A class is a blueprint or template for creating objects. It defines the properties (attributes) and behaviors (methods) of an object.

````python
# Define a class
class Car:
    def __init__(self, brand, color):
        self.brand = brand  # Attribute
        self.color = color  # Attribute

    def drive(self):  # Method
        print(f"The {self.color} {self.brand} car is driving!")

# Create an object
my_car = Car("Toyota", "Red")

# Access attributes and methods
print(my_car.brand)      # Output: Toyota
my_car.drive()           # Output: The Red Toyota car is driving!
````

# **Explanation:**

class Car: Defines a blueprint for the Car object.

__init__: A special method (constructor) to initialize object attributes.

my_car: An object of the Car class.

display(): A method to print car details.

# **Inheritance**

One object acquires the properties and behavior of the parent object.
It creates a parent-child relationship between two classes.
Benefit: Code reusability.

````python
# Parent Class
class Animal:
    def speak(self):
        print("Animal speaks")

# Child Class (inherits Animal)
class Dog(Animal):
    def speak(self):
        print("Dog barks")

# Create object of Dog
dog1 = Dog()
dog1.speak()  # Output: Dog barks
````

# **Explanation:**

Animal: Parent class (base class).

Dog: Child class (derived class) inherits from Animal.

Method Overriding: The speak() method is overridden in the child class.

# **Method Overloading(Same method name, different parameters.)**

````python
class Calculator:
    # Method to add numbers
    def add(self, a=None, b=None):
        if a is not None and b is not None:
            return a + b
        return a

calc = Calculator()
print(calc.add(5, 10))  # Output: 15
print(calc.add(5))      # Output: 5
````

# **Method Overriding(Child class changes parent method)**

````python
class Bird:
    def fly(self):
        print("Bird can fly")

class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly")

penguin = Penguin()
penguin.fly()  # Output: Penguin cannot fly
````

# **Abstraction**
Abstraction hides the implementation details and shows only the essential features.

In Python, we use the ABC module to create abstract classes.

````python
from abc import ABC, abstractmethod

# Abstract Class
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

# Concrete Class (Implements Abstract Method)
class Dog(Animal):
    def sound(self):
        print("Dog barks")

dog = Dog()
dog.sound()  # Output: Dog barks
````

# **Encapsulation**

Encapsulation is data hiding by making attributes private and using methods to access them.

````python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private attribute

    # Public method to access private attribute
    def get_balance(self):
        return self.__balance

account = BankAccount(5000)
print(account.get_balance())  # Output: 5000
````

**Explanation:**

__balance: Private attribute (cannot be accessed directly).

get_balance(): Public method to retrieve private data.

# **Write a program to count most repeating character?**

````python
from collections import Counter

s = "ajaykarre"
counter = Counter(s)

# Getting the most common character
most_common_char = counter.most_common(1)[0][0]

print(most_common_char)  # Output: 'a'

````

# **write a program Check for Balanced Brackets Using Stack?**

````python
def is_balanced(s):
    stack = []
    brackets = {')': '(', '}': '{', ']': '['}  # Mapping of closing to opening brackets

    for char in s:
        if char in brackets.values():  # If it's an opening bracket
            stack.append(char)
        elif char in brackets.keys():  # If it's a closing bracket
            if not stack or stack.pop() != brackets[char]:
                return False  # Mismatch found, return False

    return not stack  # If stack is empty, it's balanced; otherwise, unbalanced

# Test Cases
print(is_balanced("{[]}"))   # ✅ Output: True (Balanced)
print(is_balanced("{[])"))   # ❌ Output: False (Unbalanced)
print(is_balanced("(([]){})"))  # ✅ Output: True (Balanced)
print(is_balanced("{[]){"))  # ❌ Output: False (Unbalanced)
print(is_balanced("{[()]}}"))  # ❌ Output: False (Unbalanced)
````
