# **What is Logistic Regression?**

Logistic Regression is a supervised learning algorithm used for classification tasks. Despite its name, it is not a regression algorithm but is used to predict categorical outcomes (e.g., Yes/No, Spam/Not Spam, Fraud/Not Fraud).

# **what is mean by K-Nearest Neighbors K-NN?**

**The K-Nearest Neighbors (K-NN) algorithm is a simple and effective supervised learning method used for classification tasks, including brain tumor detection. It classifies a given data point based on the majority class of its K-nearest neighbors in feature space.**

KNN.jpeg

K-NN is a non-parametric classification algorithm that assigns a class label based on the majority vote of its K-nearest neighbors. The classification is performed using a distance metric to measure the similarity between data points.
